```python
# merging all the files by date - Tech Stocks
m1 = pd.merge(Aapl_tech, Amzn_tech, how="inner", left_on="Date_aapl", right_on="Date_amzn")
m2 = pd.merge(m1, Fb_tech, how="inner", left_on="Date_aapl", right_on="Date_fb")
m3 = pd.merge(m2, Goog_tech, how="inner", left_on="Date_aapl", right_on="Date_goog")
m4 = pd.merge(m3, Ibm_tech, how="inner", left_on="Date_aapl", right_on="Date_ibm")
tech_stocks = pd.merge(m4, Msft_tech, how="inner", left_on="Date_aapl", right_on="Date_msft")

tech_stocks.head()
```

```python
# merging all the files by date - Finance Stocks 
m1 = pd.merge(bcs_fin, cs_fin, how="inner", left_on="Date_bcs", right_on="Date_cs")
m2 = pd.merge(m1, db_fin, how="inner", left_on="Date_bcs", right_on="Date_db")
m3 = pd.merge(m2, gs_fin, how="inner", left_on="Date_bcs", right_on="Date_gs")
m4 = pd.merge(m3, ms_fin, how="inner", left_on="Date_bcs", right_on="Date_ms")
fin_stocks = pd.merge(m4, wfc_fin, how="inner", left_on="Date_bcs", right_on="Date_wfc")

fin_stocks.head()
```

```python
# merging all the files by date- Aviation stocks
m1 = pd.merge(aal_avn, algt_avn, how="inner", left_on="Date_aal", right_on="Date_algt")
m2 = pd.merge(m1, alk_avn, how="inner", left_on="Date_aal", right_on="Date_alk")
m3 = pd.merge(m2, dal_avn, how="inner", left_on="Date_aal", right_on="Date_dal")
m4 = pd.merge(m3, ha_avn, how="inner", left_on="Date_aal", right_on="Date_ha")
avn_stocks = pd.merge(m4, luv_avn, how="inner", left_on="Date_aal", right_on="Date_luv")

avn_stocks.head()
```

```python
# merging all the files by date- Healthcare sector
m1 = pd.merge(bhc_pharma, jnj_pharma, how="inner", left_on="Date_bhc", right_on="Date_jnj")
m2 = pd.merge(m1, mrk_pharma, how="inner", left_on="Date_bhc", right_on="Date_mrk")
m3 = pd.merge(m2, pfe_pharma, how="inner", left_on="Date_bhc", right_on="Date_pfe")
m4 = pd.merge(m3, rhhby_pharma, how="inner", left_on="Date_bhc", right_on="Date_rhhby")
pharma_stocks = pd.merge(m4, unh_pharma, how="inner", left_on="Date_bhc", right_on="Date_unh")

pharma_stocks.head()
```

```python
# Subsetting only the closing prices column for plotting - Tech Industry
tech_stocks['Date']= pd.to_datetime(tech_stocks['Date_aapl'],dayfirst = True)
tech_close = tech_stocks[["Date","Close_aapl", "Close_amzn","Close_goog","Close_ibm", "Close_msft", "Close_fb"]]
tech_close.head()
```

```python
# Subsetting only the closing prices column for plotting - Finance Industry
fin_stocks['Date']= pd.to_datetime(fin_stocks['Date_bcs'],dayfirst = True)
fin_close = fin_stocks[["Date","Close_bcs", "Close_cs", 'Close_db', "Close_gs","Close_ms", "Close_wfc"]]
fin_close.head()
```

```python
# Subsetting only the closing prices column for plotting - Aviation Industry
avn_stocks['Date']= pd.to_datetime(avn_stocks['Date_aal'],dayfirst = True)
avn_close = avn_stocks[["Date","Close_aal", "Close_algt", 'Close_alk', "Close_dal","Close_ha", "Close_luv"]]
avn_close.head()
```

```python
# Subsetting only the closing prices column for plotting - Pharma Industry
pharma_stocks['Date']= pd.to_datetime(pharma_stocks['Date_bhc'],dayfirst = True)
pharma_close = pharma_stocks[["Date","Close_bhc", "Close_jnj", 'Close_mrk', "Close_pfe","Close_rhhby", "Close_unh"]]
pharma_close.head()
```

```python
# Sort the stock data by date - Tech Stocks 
tech_df = tech_close.sort_values(by = ['Date'])
tech_df = tech_df.dropna()
tech_df.head()
```

```python
# Sort the stock data by date - Finance Stocks 
fin_df = fin_close.sort_values(by = ['Date'])
fin_df = fin_df.dropna()
fin_df.head()
```

```python
# Sort the stock data by date - Aviation Stocks 
avn_df = avn_close.sort_values(by = ['Date'])
avn_df = avn_df.dropna()
avn_df.head()
```

```python
# Sort the stock data by date - Pharma Stocks 
pharma_df = pharma_close.sort_values(by = ['Date'])
pharma_df = pharma_df.dropna()
pharma_df.head()
```
